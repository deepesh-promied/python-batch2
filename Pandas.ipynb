# Subsetting Rows

```python
df[df['Marks'] > 80] # Marks greater than 80
```

```python
df['Subject']=='Chemistry'
```

```python
df[df['Subject']=='Chemistry']
```

```python
df[(df['Subject']=='Chemistry') & (df['Marks'] >80)] # Multiple Conditions
```

```python
cond1 = df['Subject']=='Chemistry'
cond2 = df['Marks'] >80
df[cond1 & cond2]
```

```python
cond3 = df['Subject']=='History'
df[(cond1 | cond3) & cond2]
```
